## Function to extract the User Summary

```python
def get_user_summary(user_id):
    url = f"{base_url}/u/{user_id}/summary.json"

    headers = {
        "Api-Key": API_KEY,
        "Api-Username": API_USERNAME
    }

    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        return {"error": f"Failed to fetch data for user_id: {user_id}", "status_code": response.status_code}

    data = response.json()
    return data
```

## Extract the Basic Metrics

```python
user_id = "sayan"
summary_data = get_user_summary(user_id)

user_summary = summary_data.get("user_summary", {})

fields_to_show = [
    "likes_given", "likes_received", "topics_entered", "posts_read_count", "days_visited",
    "topic_count", "post_count", "time_read", "recent_time_read",
    "can_see_summary_stats", "can_see_user_actions", "solved_count"
]

filtered_summary = {key: user_summary[key] for key in fields_to_show if key in user_summary}

df = pd.DataFrame(filtered_summary.items(), columns=["Metric", "Value"])
df.style.set_properties(**{'text-align': 'left'}).set_table_styles(
    [{'selector': 'th', 'props': [('text-align', 'left')]}]
)
```

## Function to get information how much this student was active in the top subjects.

```python
top_categories = summary_data.get("user_summary", {}).get("top_categories", [])

category_summary = [
    {
        "Course Name": category["name"],
        "Topics Created": category["topic_count"],
        "Posts Replied": category["post_count"]
    }
    for category in top_categories
]

df_categories = pd.DataFrame(category_summary)

df_categories
```
